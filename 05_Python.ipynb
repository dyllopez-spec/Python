# Ejercicio 3
Genera 100 números aleatorios entre 1 y 100 e indica cuántos fueron pares y cuántos
impares.


```python
import random 
contador_par=0
contador_impar=0
for i in range(100):
    num = (random.randint (1,100))
    print(f"{i}:{num}")
    if num % 2==0:
        contador_par+=1  
    elif num % 2==1:
       contador_impar+=1

print (f"los numeros pares son: {contador_par}")   
print (f"los numeros impares son: {contador_impar}")
```

# Ejercicio 4
Genera 40 números aleatorios entre 1 y 200 y suma únicamente los que sean múltiplos de
5

```python
import random
suma_total=0
for i in range (40):
    num=(random.randint (1,200))
    print(f"{i}:{num}")
    if num%5==0:
        suma_total+=num
print (f"la totalida de los numeros divisibles en 5 es: {suma_total}")


```

# Ejercicio 5
Genera 30 temperaturas aleatorias entre -5 y 35 grados.
Luego muestra la suma total, cuántas fueron bajo 0 grados, cuántas fueron mayores a 25
grados y el promedio general usando acumulador y contadores.


```python
import random 
contador_zero=0
acumulador_temp25=0
total=0
cantidad_temp=30
for i in range (30):
    temperatura= (random.randint(-5,35))
    print(temperatura)
    total+=temperatura 
    if temperatura < 0:
        contador_zero+=1
    elif temperatura > 25:
        acumulador_temp25+=1

promedio=(total/cantidad_temp)


print(f"los grados menores a 0 son: {contador_zero}")
print(f"los grados mayores a 25 son: {acumulador_temp25}")
print (f"el promeddio de la temperatura es: {promedio}")


```
