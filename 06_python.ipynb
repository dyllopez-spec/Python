# Contexto:
En una feria tecnológica se realizará un juego llamado “Panel de Energía”. Cada
participante debe ingresar varios valores numéricos, y el sistema analizará los datos
para entregar un resultado final según ciertas reglas. El objetivo es practicar
decisiones, repeticiones, validaciones y cálculos acumulados en Python.

# Objetivo:
Crear un programa en Python que solicite una cantidad determinada de rondas y, en
cada ronda, pida al usuario un número entero positivo. Luego, el programa debe
clasificar cada número, llevar estadísticas adicionales y mostrar un resumen final
más completo

# Requisitos
El programa debe:
* Pedir al usuario cuántas rondas desea jugar.
* Validar que la cantidad de rondas sea mayor que cero.
* Usar un ciclo para repetir el proceso esa cantidad de veces.
* En cada ronda, solicitar un número entero positivo.
* Validar que el número sea mayor que cero; si no lo es, volver a pedirlo.

# Clasificación
En cada ronda, el número debe clasificarse así:
* Si es múltiplo de 2 y de 3, mostrar “Energía alta”.
* Si solo es múltiplo de 2, mostrar “Energía par”.
* Si solo es múltiplo de 3, mostrar “Energía triple”.
* En cualquier otro caso, mostrar “Energía simple”.
Además de lo que ya tienes, el programa debe:
* Contar cuántos números son pares e impares.
* Contar cuántos números son mayores que 50.
* Sumar por separado los números pares y los impares.
* Indicar en cada ronda si el número ingresado es:
* pequeño, si está entre 1 y 20;
* mediano, si está entre 21 y 50;
* grande, si es mayor que 50.
* Permitir terminar antes el juego si el usuario ingresa un número especial, por
ejemplo 666, usando una condición de corte.
* Mostrar cuántas rondas realmente se jugaron si el juego terminó antes.



# Resumen final
Al finalizar, el programa debe mostrar:
* Total de números válidos ingresados.
* Cantidad de números en cada categoría de energía.
* Cantidad de números pares e impares.
* Cantidad de números mayores que 50.
* Número mayor ingresado.
* Número menor ingresado.
* Suma total de todos los números.
* Suma de números pares.
* Suma de números impares.
*  Promedio general de los números ingresados.
