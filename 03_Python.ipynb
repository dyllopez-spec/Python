
# Ejercicio 1
Solicita al usuario que ingrese una frase. Muestra cuántos caracteres tiene la frase, pero sin
contar los espacios en blanco.


# Ejercicio 2
Pide al usuario su nombre completo. Luego, muestra el nombre ingresado de tres formas
diferentes: todo en mayúsculas, todo en minúsculas, y con la primera letra de cada palabra
en mayúscula.


# Ejercicio 3
Solicita una palabra o frase que pueda tener espacios al inicio o al final. Elimina esos
espacios y luego verifica si la palabra "Python" aparece dentro del texto limpio. Muestra la
posición donde comienza dicha palabra, o un mensaje indicando que no se encontró.


# Ejercicio 4
Pide al usuario que ingrese una oración. Luego, reemplaza todas las vocales 'a' por '@', las
'e' por '3', las 'i' por '1', las 'o' por '0' y las 'u' por 'U'. Muestra la oración modificada.


# Ejercicio 5
Solicita una palabra. Extrae e imprime los caracteres que se encuentran en las posiciones 2,
4 y 6 (si existen). Si la palabra es muy corta, muestra un mensaje indicándolo.


# Ejercicio 6
Pide al usuario que ingrese su correo electrónico. Luego, verifica si el correo contiene el
símbolo '@' y un punto '.' después del '@'. Muestra un mensaje indicando si el formato
parece válido o no.


# Ejercicio 7
Solicita una frase y una palabra. Cuenta cuántas veces aparece esa palabra dentro de la
frase (sin importar mayúsculas o minúsculas). Muestra el resultado.


# Ejercicio 8
Pide al usuario que ingrese una cadena de texto. Luego, crea una nueva cadena que
contenga solo los caracteres en posiciones pares (0, 2, 4, ...) de la cadena original. Muestra
ambas cadenas.


# Ejercicio 9
Solicita una frase. Invierte el orden de las palabras (no de los caracteres). Por ejemplo,
"Hola mundo Python" debería convertirse en "Python mundo Hola". Muestra el resultado.


# Ejercicio 10
Pide al usuario que ingrese un texto. Luego, elimina todos los caracteres que no sean letras
(es decir, elimina números, signos de puntuación, espacios, etc.) y muestra el texto
resultante. Por ejemplo, "¡Hola, mundo! 123" quedaría como "Holamundo".
