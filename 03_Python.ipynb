
# Ejercicio 1
Solicita al usuario que ingrese una frase. Muestra cuántos caracteres tiene la frase, pero sin
contar los espacios en blanco.


```python
#Solicita una frase al usuario
frase = 'Aprendiendo fundamentos de la programacion'
cantida_original = len(frase)
print(f'La cantidad de caracteres de la frase original es: {cantida_original}')
#Contar la cantidad de caracteres de la fase, no se deben contar los espacios en blanco.
sin_espacios = frase.replace(' ', '')
cantidad_caracteristicas = len(sin_espacios)
print(f'La cantidad de caracteres de la frase sin espacios son: {cantidad_caracteristicas}')
```

# Ejercicio 4

'e' por '3', las 'i' por '1', las 'o' por '0' y las 'u' por 'U'. Muestra la oración modificada.

# Ejercicio 6
Pide al usuario que ingrese su correo electrónico. Luego, verifica si el correo contiene el
símbolo '@' y un punto '.' después del '@'. Muestra un mensaje indicando si el formato
parece válido o no.


```python
# Solicita un correo electronico al usuario
email = 'platon@academia.gr'
# Verifica si el correco contiene el simbolo @ y punto (.)
verifica_arroba = email.find('@')
verifica_punto = email.find('.')

if verifica_arroba != -1 and verifica_punto != -1:
# Muestra el mensaje si el formato parece valido o no
    print(f'El formato del email {email} es valido')
else:
    print(f'El formato de {email} no es valido')
```
