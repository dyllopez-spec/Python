
# Ejercicio 1
Solicita al usuario que ingrese una frase. Muestra cuántos caracteres tiene la frase, pero sin
contar los espacios en blanco.


```python
#Solicita una frase al usuario
frase = 'Aprendiendo fundamentos de la programacion'
cantida_original = len(frase)
print(f'La cantidad de caracteres de la frase original es: {cantida_original}')
#Contar la cantidad de caracteres de la fase, no se deben contar los espacios en blanco.
sin_espacios = frase.replace(' ', '')
cantidad_caracteristicas = len(sin_espacios)
print(f'La cantidad de caracteres de la frase sin espacios son: {cantidad_caracteristicas}')
```

# Ejercicio 2
Pide al usuario su nombre completo. Luego, muestra el nombre ingresado de tres formas
diferentes: todo en mayúsculas, todo en minúsculas, y con la primera letra de cada palabra
en mayúscula.


```python
#Solicitar un nombre completo
nombre_completo = 'Ada Lovelace'
#Mostrar el nombre de tres formas distintas 
#El nomnbre se mostrar en formato mayusculo
nombre_mayuscula = nombre_completo.upper()
print(f'EL nombre con Mayusculas {nombre_mayuscula}')
#El nomnbre se mostrar en formato minuscula
nombre_minuscula = nombre_completo.lower()
print(f'EL nombre con Mayusculas {nombre_minuscula}')
#El nomnbre se mostrar en formato con la priomera palabra mayuscula
nombre_capitalize = nombre_completo.title()
print(f'EL nombre con Mayusculas {nombre_capitalize}')
```

# Ejercicio 3
Solicita una palabra o frase que pueda tener espacios al inicio o al final. Elimina esos
espacios y luego verifica si la palabra "Python" aparece dentro del texto limpio. Muestra la
posición donde comienza dicha palabra, o un mensaje indicando que no se encontró.


```python
#SOlicitar una frase o nombre al ususario
frase = '  Aprendiendo a programar con Python '
# Elimina  los espacios vacios
sin_espacios = frase.strip()
#Busca la posicion de la palabra
verifica = sin_espacios.find('Python')
#Muestra la posicion de la palabra
if (verifica):
    print(f'La posicion de la palabra es {verifica}')
else:
    print(f'La palabra "Python" no se encuentra')
```
